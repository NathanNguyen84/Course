Question 5 - Tesla Stock and Revenue Dashboard 

```python
import matplotlib.pyplot as plt

# Fetch the Tesla stock data
stock = 'TSLA'
ticker = yf.Ticker(stock)
ticker_ = ticker.ticker 
data = yf.download(ticker_, start="2021-01-01", end="2023-12-31")


# Fetch the financials for Tesla
financials = ticker.financials
revenue_data = financials.loc['Total Revenue']

# Create the dashboard
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot the stock data
ax1.plot(data['Close'])
ax1.set_title('Tesla Stock Data')
ax1.set_ylabel('Price')

# Plot the revenue data
ax2.plot(revenue_data)
ax2.set_title('Tesla Revenue')
ax2.set_ylabel('Revenue')

# Adjust the layout
plt.tight_layout()

# Show the dashboard
plt.show()

```
