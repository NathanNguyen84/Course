```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates
```

```python
# Load historical stock price data into a Pandas DataFrame
data = pd.read_csv('stock_MSFT.csv')
```

```python
data= data.tail (600)
# Calculate the 50-day simple moving average (SMA)
data['SMA_L'] = data['Adj Close'].rolling(window=60).mean()
data['SMA_M'] = data['Adj Close'].rolling(window=20).mean()
data['SMA_S'] = data['Adj Close'].rolling(window=5).mean()

# Plot the stock price and SMA

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)


ax1.plot(data['Date'], data['Adj Close'], label='Stock Price')
ax1.plot(data['Date'], data['SMA_L'], label='SMA (60 days)')
ax1.plot(data['Date'], data['SMA_M'], label='SMA (20 days)')
ax1.plot(data['Date'], data['SMA_S'], label='SMA (5 days)')
ax1.set_xlabel('Date')
ax1.set_ylabel('Price')
ax1.set_title('Stock Price with SMA')
ax1.legend()

# Add volume to the chart

ax2.bar(data['Date'], data['Volume'], color='blue')
ax2.set_ylabel('Volume')
ax2.set_xlabel('Date')

plt.tight_layout()
plt.show()
```
